```python
#Mini Project: Geometry Cakculator

def calculate_area(length, width):
    return length * width

def calculate_perimeter(length, width):
    return 2*(length+width)

length=int(input("Enter length: "))
width=int(input("Enter width: "))

while True:
    task=input("type a to calculate area, p to calculate perimeter or q to quit: ")
    if task=='a':
        area=calculate_area(length, width)
        print(f"Area of rectangle with length {length} and width {width} is {area}")
    elif task=='p':
        perimeter=calculate_perimeter(length, width)
        print(f"Perimeter of rectangle with length {length} and width {width} is {perimeter}")
    elif task=='q':
        break
    else:
        print("invalid input")
        continue
```

```python
#Classes and Objects

class testClass: #blueprint
    x=5
obj1=testClass() #instance
print(obj1.x)

class human:
    def __init__(self, name, age=18):
        self.name=name
        self.age=age
obj2=human("swayam",21)
print(obj2.name, obj2.age)

class car:
    def __init__(self, brand, model, color):
        self.brand=brand
        self.model=model
        self.color=color
    def identify(self):
        return f"The {self.color} colored car is a {self.brand} {self.model}"
obj3=car("Ford","Mustang","Red")
print(obj3.identify())

class person:
    def __init__(self, fname, lname):
        self.firstname=fname
        self.lastname=lname
    def greet(self):
        return f"Hello {self.firstname} {self.lastname}!"
    def welcome(self):
        greeting=self.greet()
        return f"{greeting} Welcome!"
obj4=person("Swayam","Parekh")
print(obj4.welcome())

class intern(person):
    def __init__(self, fname, lname,batch):
        super().__init__(fname, lname)
        self.batch=batch
    def welcome(self):
        greeting=self.greet()
        return f"{greeting} Welcome to the {self.batch} batch!"
obj5=intern("Swayam","Parekh",2026)
print(obj5.welcome())
```

```python
#Mini project: Bank Account System
class BankAccount:
    balance=0
    def __init__(self, account_holder):
        self.account_holder=account_holder
    def deposit(self, amount):
        self.balance+=amount
        print(f"₹{amount} deposited. New balance is ₹{self.balance}")
    def withdraw(self, amount):
        if amount>self.balance:
            print("Insufficient balance")
        else:
            self.balance-=amount
            print(f"₹{amount} withdrawn. New balance is ₹{self.balance}")
    def check_balance(self):
        print(f"Current balance is ₹{self.balance}")
account=BankAccount("Swayam Parekh")
account.deposit(1000)
account.withdraw(500)
account.check_balance()
account.withdraw(600)
```
