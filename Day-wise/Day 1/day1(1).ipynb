```python
#Descriptive Statistics
#Mini Project: Stats Calculator
a=[1,2,3,4,5,6,7,8,8,9,10,11,12,13]
mean=sum(a)/len(a)
print("mean: ", mean)
if len(a)%2==0:
    print("median: ", (sorted(a)[len(a)//2-1]+sorted(a)[len(a)//2])/2)
else:
    print("median: ", sorted(a)[len(a)//2])
print("mode: ", max(set(a), key=a.count))

#variance
variance=sum((x-mean)**2 for x in a)/len(a)
print("variance: ", variance)

print("standard deviation: ", variance**0.5)
```

```python
import numpy as np

array_1d=np.array([1,2,3,4])
array_2d=np.array([[1,2,3],[4,5,6]])
array_3d=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
zeroes=np.zeros((5,5))
ones=np.ones(10)
print("1D: ",array_1d)
print("\n2D: ",array_2d)
print("\n3D: ",array_3d)
print("\nZeroes: ",zeroes)
print("\nOnes: ",ones)

a1=np.array([1,2,3,4,5])
a2=np.array([6,7,8,9,10])
print("vectorized addition",a1+a2)

vec=np.vectorize(lambda x: x**2)
a3=vec(a1)
print("vectorized custom function",a3)

print("Indexing: ", array_2d[1,2])  
print("Slicing: ", array_2d[:,1])
print("Boolean indexing: ", array_2d[array_2d<3])
```

```python
#Mini Project: Titanic Analysis

df=pd.read_csv("Titanic-Dataset.csv")
average_age=df["Age"].mean()
print("Average age of passengers: ",average_age)
print("-------------------------------")
print(df.isnull().sum())
print("-------------------------------")
df["Age"].fillna(df["Age"].median(), inplace=True)
print(df.isnull().sum())
print("-------------------------------")
df1=df[(df["Survived"]==1) & (df["Sex"]=="female")]
print(df1.head())
print("-------------------------------")
df2=df.groupby("Pclass")["Survived"].mean()
print(df2)
```
